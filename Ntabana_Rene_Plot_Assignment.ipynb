```python
# YOUR CODE HERE
binwidth = 60
plt.hist(fuel_econ.co2, edgecolor="black", bins=range(min(fuel_econ.co2), max(fuel_econ.co2) + binwidth, binwidth))
plt.xticks(range(30, max(fuel_econ.co2) + binwidth, binwidth))
plt.xlabel("CO2")
plt.ylabel("Frequency")
plt.show()
```

```python
# YOUR CODE HERE
plt.hist(fuel_econ.drive)
plt.xticks(rotation=60)
plt.ylabel("Frequency")
plt.xlabel("Drive")
plt.show()
```

**Response for the question:** Finally What Columns would you delete Drop if the target is the ghg score? Justify your Answer

I would drop columns that are highly correlated to avoid multicollinearity. As it can be seen on the the heatmap diagram, columns 'city', 'highway', 'comb', 'co2', 'feScore' and 'UHighway' are relatively highly correlated with correlation value ranging between + or - 0.9 and + or - 1.

The following script shows the heatmap digram after dropping the above columns.


```python
new_fuel_econ = fuel_econ.drop(columns=['city', 'highway', 'comb', 'co2', 'feScore', 'UHighway'])
sb.heatmap(new_fuel_econ.corr(), cmap="RdBu")
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```
